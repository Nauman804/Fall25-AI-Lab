```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt

from sklearn.model_selection import train_test_split
from sklearn.naive_bayes import BernoulliNB, GaussianNB, MultinomialNB
from sklearn.ensemble import RandomForestClassifier
from sklearn.tree import DecisionTreeClassifier
from sklearn.neighbors import KNeighborsClassifier

from sklearn import metrics
from sklearn.metrics import accuracy_score
```

```python
x = data.iloc[:, 0:-1]  
y = data.iloc[:, -1]     
```

```python
x[cat_columns] = x[cat_columns].apply(lambda col: pd.factorize(col)[0])
```

```python
X_train, X_test, Y_train, Y_test = train_test_split(
    x, y, test_size=0.3, shuffle=False
)
```

```python
BernNB = BernoulliNB()
BernNB.fit(X_train, Y_train)
Y_bpred = BernNB.predict(X_test)

b_accuracy = metrics.accuracy_score(Y_test, Y_bpred)
b_precision = metrics.precision_score(Y_test, Y_bpred, average='weighted', zero_division=0)
b_recall = metrics.recall_score(Y_test, Y_bpred, average='weighted', zero_division=0)
b_f1 = metrics.f1_score(Y_test, Y_bpred, average='weighted', zero_division=0)

print("BernoulliNB Scores:")
print("Accuracy:", b_accuracy, "Precision:", b_precision, "Recall:", b_recall, "F1:", b_f1)
```

```python
RF = RandomForestClassifier()
RF.fit(X_train, Y_train)
Y_rpred = RF.predict(X_test)

r_accuracy = accuracy_score(Y_test, Y_rpred)
r_precision = metrics.precision_score(Y_test, Y_rpred, average='weighted', zero_division=0)
r_recall = metrics.recall_score(Y_test, Y_rpred, average='weighted', zero_division=0)
r_f1 = metrics.f1_score(Y_test, Y_rpred, average='weighted', zero_division=0)

print("Random Forest Scores:")
print("Accuracy:", r_accuracy, "Precision:", r_precision, "Recall:", r_recall, "F1:", r_f1)

```

```python
GausNB = GaussianNB()
GausNB.fit(X_train, Y_train)
Y_gpred = GausNB.predict(X_test)

g_accuracy = accuracy_score(Y_test, Y_gpred)
g_precision = metrics.precision_score(Y_test, Y_gpred, average='weighted', zero_division=0)
g_recall = metrics.recall_score(Y_test, Y_gpred, average='weighted', zero_division=0)
g_f1 = metrics.f1_score(Y_test, Y_gpred, average='weighted', zero_division=0)

print("GaussianNB Scores:")
print("Accuracy:", g_accuracy, "Precision:", g_precision, "Recall:", g_recall, "F1:", g_f1)

```

```python
Dtree = DecisionTreeClassifier()
Dtree.fit(X_train, Y_train)
Y_dpred = Dtree.predict(X_test)

d_accuracy = accuracy_score(Y_test, Y_dpred)
d_precision = metrics.precision_score(Y_test, Y_dpred, average='weighted', zero_division=0)
d_recall = metrics.recall_score(Y_test, Y_dpred, average='weighted', zero_division=0)
d_f1 = metrics.f1_score(Y_test, Y_dpred, average='weighted', zero_division=0)

print("Decision Tree Scores:")
print("Accuracy:", d_accuracy, "Precision:", d_precision, "Recall:", d_recall, "F1:", d_f1)

```

```python
MultiNB = MultinomialNB()
MultiNB.fit(X_train, Y_train)
Y_mpred = MultiNB.predict(X_test)

m_accuracy = accuracy_score(Y_test, Y_mpred)
m_precision = metrics.precision_score(Y_test, Y_mpred, average='weighted', zero_division=0)
m_recall = metrics.recall_score(Y_test, Y_mpred, average='weighted', zero_division=0)
m_f1 = metrics.f1_score(Y_test, Y_mpred, average='weighted', zero_division=0)

print("MultinomialNB Scores:")
print("Accuracy:", m_accuracy, "Precision:", m_precision, "Recall:", m_recall, "F1:", m_f1)
```

```python
KNN = KNeighborsClassifier()
KNN.fit(X_train, Y_train)
Y_kpred = KNN.predict(X_test)

k_accuracy = accuracy_score(Y_test, Y_kpred)
k_precision = metrics.precision_score(Y_test, Y_kpred, average='weighted', zero_division=0)
k_recall = metrics.recall_score(Y_test, Y_kpred, average='weighted', zero_division=0)
k_f1 = metrics.f1_score(Y_test, Y_kpred, average='weighted', zero_division=0)

print("KNN Scores:")
print("Accuracy:", k_accuracy, "Precision:", k_precision, "Recall:", k_recall, "F1:", k_f1)
```

```python
plt.figure(figsize=(16,10))

models = ['Bernoulli','Random Forest','Gaussian','Decision Tree','Multinomial','KNeighbors']

plt.plot(models, [b_accuracy, r_accuracy, g_accuracy, d_accuracy, m_accuracy, k_accuracy], marker='o', label='Accuracy')
plt.plot(models, [b_precision, r_precision, g_precision, d_precision, m_precision, k_precision], marker='o', label='Precision')
plt.plot(models, [b_recall, r_recall, g_recall, d_recall, m_recall, k_recall], marker='o', label='Recall')
plt.plot(models, [b_f1, r_f1, g_f1, d_f1, m_f1, k_f1], marker='o', label='F1 Score')

plt.title("Scores of Applied Classifiers")
plt.xlabel("Classifiers")
plt.ylabel("Score")
plt.ylim(0,1.05)
plt.legend()
plt.grid(True)
plt.show()
```
